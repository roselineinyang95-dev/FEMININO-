# Extracting order hour from order date variable

# Creating time of day variable

```python
def get_time_of_day(hour):
    if  0 <= hour < 4:
        return "Dawn"
    elif 7 < hour < 12:
        return "Morning"
    elif 12 < hour < 16:
        return "Afternoon"
    elif 16 < hour < 20:
        return "Evening"
    else:
        return "Night"

Customer_order_feedback["time_of_day"] = Customer_order_feedback["order_hour"].apply(get_time_of_day)
##Creating time of day variable
```

```python
subset = Customer_order_feedback[["time_of_day", "order_date"]]
(subset)
```

# Checking the time of day orders was most made

```python
Time_analysis = Customer_order_feedback.groupby(["time_of_day"])["order_id"].count().reset_index()
Time_analysis.rename(columns={'order_id':'order_count'},inplace=True)  #renaming variable for clarity

Time_analysis= Time_analysis.sort_values(by='order_count',ascending=False).reset_index()
(Time_analysis)
```

# Creating 'Satisfaction table' to be able to analyse its probability

```python
# creatinng satisfaction table
def get_satisfaction_table(rating):
    if rating == 1:
        return "Very Poor"
    elif rating == 2:
        return "Poor"
    elif rating == 3:
        return "Neutral"
    elif rating == 4:
        return "Satisfied"
    elif rating == 5:
        return "Very Satisfied"
    else:
        return "Unknown"

Customer_order_feedback["satisfaction_table"] = Customer_order_feedback["rating"].apply(get_satisfaction_table)
```

```python
# 1. Create a pivot table for counts of each rating by customer segment
rating_counts = pd.crosstab(
    Customer_order_feedback["customer_segment"],
    Customer_order_feedback["rating"]
)

# 2. Rename the columns so they look like rating_1, rating_2, etc.
rating_counts = rating_counts.rename(
    columns={i: f"rating_{int(i)}" for i in rating_counts.columns}
)

# 3. Add average rating for each segment
avg_ratings = (
    Customer_order_feedback.groupby("customer_segment")["rating"]
    .mean()
    .rename("avg_rating")
)

# 4. Combine into one table
segment_summary = rating_counts.join(avg_ratings)

segment_summary
```
